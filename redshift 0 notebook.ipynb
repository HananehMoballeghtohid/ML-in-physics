```python
import numpy as np
import pandas as pd
```

```python
from lenstools.simulations import Gadget2SnapshotDE , Gadget2Snapshot
from astropy.units import Mpc, m, s
```

```python
snap = Gadget2Snapshot.open('lcdmML_snap003_cdm')
```

```python
pos_0 = snap.getPositions()
```

```python
pos_0.info()
```

```python
velocities_0 = snap.getVelocities()
```

```python
velocities_0.info()
```

```python
ids_0 = snap.getID()
```

```python
ids_0
```

```python
ids_0.shape
```

```python
df = pd.DataFrame(ids_0, columns= ['ids_0'])
df_1 = pd.DataFrame(pos_0, columns= ['x_0','y_0','z_0'])
df_2 = pd.DataFrame(velocities_0, columns= ['v_x_0','v_y_0','v_z_0'])
df[df_1.columns] = df_1[df_1.columns]
df[df_2.columns] = df_2[df_2.columns]
df
```

```python
#it shows our dataframe has no duplicate
df = df.drop_duplicates()
df
```

```python
len(df['ids_0'].unique())
```

```python
#numbers show that all repeated ids have value '0'. so if we remove all '0's from ids_0 column all ids would be unique
df[df['ids_0'] == 0]
```

```python
df_uniques = df[df['ids_0'] != 0]
df_uniques
```

```python
#showing correlation between features
df.corr()
```

```python
#showing correlation between features
df_uniques.corr()
```

```python
df.to_csv('RedSift0_with_ids', index = False)
```
