```python
import numpy as np
import pandas as pd
```

```python
from lenstools.simulations import Gadget2SnapshotDE , Gadget2Snapshot
from astropy.units import Mpc, m, s
```

```python
snap = Gadget2Snapshot.open('lcdmML_snap000_cdm')
```

```python
pos_1 = snap.getPositions()
```

```python
pos_1.info()
```

```python
velocities_1 = snap.getVelocities()
```

```python
velocities_1.info()
```

```python
ids_1 = snap.getID()
```

```python
ids_1
```

```python
ids_1.shape
```

```python
df = pd.DataFrame(ids_1, columns= ['ids_100'])
df_1 = pd.DataFrame(pos_1, columns= ['x_100','y_100','z_100'])
df_2 = pd.DataFrame(velocities_1, columns= ['v_x_100','v_y_100','v_z_100'])
df[df_1.columns] = df_1[df_1.columns]
df[df_2.columns] = df_2[df_2.columns]
df
```

```python
#it shows our dataframe does'nt have duplicates
df = df.drop_duplicates()
df
```

```python
len(df['ids_100'].unique())
```

```python
"""numbers show that all repeated values in 'ids_0' column are in fact 0 so if we remove them we would 
have a dataframe with unique values"""

df[df['ids_100'] == 0]
```

```python
df_uniques = df[df['ids_100'] != 0]
df_uniques
```

```python
#showing correlation between features
df.corr()
```

```python
#showing correlation between features
df_uniques.corr()
```

```python
df.to_csv('RedSift100_with_ids', index = False)
```
